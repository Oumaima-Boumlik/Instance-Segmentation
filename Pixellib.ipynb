```
import pixellib
from pixellib.instance import instance_segmentation
#We imported in the class for performing semantic segmentation and created an instance of the class.
segment_video = instance_segmentation()
```

```
#We loaded the xception model trained on pascal voc dataset to perform semantic segmentation
segment_video.load_model("mask_rcnn_coco.h5")
```

```
import time
startx = time.time()
segment_video.process_video("Comp-1.mp4", show_bboxes = True, frames_per_second= 15, output_video_name="pixellib_instanceSeg_pred.mp4")
pred_speed = time.time() - startx
print('Prediction Speed (fps): ', pred_speed/631 ) #nbr of frames
```
